# Challenge 3 - Constructing Confidence Intervals

We will test whether the hourly wage of all hourly workers is significantly different from $30/hr.

In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. Is $30/hr within that interval?

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python

# Your code here
import scipy.stats as st  
# we want to get where " ['Salary or Hourly']" is Horly , we choice hourly
hourly_workers = salaries [salaries ['Salary or Hourly'] == 'Hourly'].copy()

#Here we are gonna to take the column that filter by hourly and hourly rate y creamos "hourly rate" column#
hourly_data = hourly_workers['Hourly Rate'].dropna() 

# Get the Confidencial Interval for all Hourly Workers = (medimos la columna que hemos creado precios por hora)
# Required parameters 
conf_level = 0.95
degrees_f = len (hourly_data) -1
#calculamos el promedio de esos precios
mean_workers = hourly_data.mean()

st_sem = st.sem(hourly_data)  #Standard Error of the Mean.

# IMPORTANTE: Guardamos en una variable con nombre distinto

st_t_interval = st.t.interval(conf_level, degrees_f, loc= mean_workers, scale= st_sem )

#ci_all = st.t.interval(
    #conf_level, # El porcentaje de seguridad (ej. 0.95)
    #df_all, o     # Grados de libertad (tamaño del grupo - 1)
    #loc=mean_all, (degrees_f )   # ¿Dónde está el centro? (Promedio)
    #scale=se_all o (scale= st_sem)   # ¿Qué tan ancha es la zona de error? (Error estándar) #

result = st_t_interval = st.t.interval(conf_level, degrees_f, loc= mean_workers, scale= st_sem )


print(f"95% Confidence Interval (All): {result}")
print(f"Is $30 inside the interval? {'Yes' if result[0] <= 30 <= result[1] else 'No'}")


print(hourly_data.head())
print(f"Total de datos: {len(hourly_data)}")

#¿Qué número te sale al imprimir len(hourly_data)? Si es 0, podemos revisar el código donde creaste esa variable.





```
